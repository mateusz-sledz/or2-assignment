```python
import math
import numpy as np
import matplotlib.pyplot as plt


def gradient(x, y):
    return [3*y*x**2 - y**3, x**3 - 3*x*y**2]


while True:
    string = input("Enter two components of point in R2 separated ONLY by space: ")

    point = string.split()
    if len(point) != 2:
        print('Input format is wrong, please specify exactly two numbers.')
        continue

    try:
        x, y = float(point[0]), float(point[1])
    except(ValueError):
        print('Non numerical values provided, try again.')
        continue
    
    break

while True:
    string = input("Enter two components of vector 'u' separated ONLY by space: ")

    vector = string.split()
    if len(vector) != 2:
        print('Input format is wrong, please specify exactly two numbers.')
        continue

    try:
        u1, u2 = float(vector[0]), float(vector[1])
    except(ValueError):
        print('Non numerical values provided, try again.')
        continue
    
    break

size = math.sqrt(u1**2 + u2**2)

u1 = u1/size
u2 = u2/size

grad = gradient(x, y)
derivative = np.array(grad) @ np.array([u1, u2])
print("Directional derivative at point ", (x, y)," in direction ", (u1, u2) ," equals: ", derivative)
print("Gradient in this point is equal: ", grad )

def fun(x, y):
    return y*x**3 - x*y**3

fig = plt.figure(figsize=(20, 15))
ax = fig.add_subplot(111, projection='3d')
x1 = np.arange(-15.0 + x, 15.0 + x , 0.1)
y1 = np.arange(-15.0 + y, 15.0 + y , 0.1)
X, Y = np.meshgrid(x1, y1)
zs = np.array([fun(x1,y1) for x1,y1 in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z, alpha=0.55)
ax.scatter3D(x, y, fun(x, y), color='r')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('F value')
plt.show()
```
